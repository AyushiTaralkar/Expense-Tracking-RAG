```python
import pandas as pd

df = pd.read_csv(r"C:\Users\ayush\Health Hack\Expense-Tracking-RAG\data\Personal_Finance_Dataset.csv")

df["Date"] = pd.to_datetime(df["Date"])
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
```

```python
total_spent = df[df["Type"] == "Expense"]["Amount"].sum()
print(total_spent)
```

```python
category_spending = (
    df[df["Type"] == "Expense"]
    .groupby("Category")["Amount"]
    .sum()
    .sort_values(ascending=False)
)
print(category_spending)
```

```python
monthly_spending = (
    df[df["Type"] == "Expense"]
    .groupby(["Year", "Month"])["Amount"]
    .sum()
)
print(monthly_spending)
```

```python
avg_monthly = monthly_spending.mean()
print(avg_monthly)
```

```python
top_category = category_spending.idxmax()
print(top_category)
```
