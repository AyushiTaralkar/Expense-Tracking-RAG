```python
import pandas as pd

df = pd.read_csv(r"C:\Users\ayush\Health Hack\Expense-Tracking-RAG\data\Personal_Finance_Dataset.csv")

df["Date"] = pd.to_datetime(df["Date"])
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
```

```python
expenses = df[df["Type"] == "Expense"]

print("Data Loaded Successfully")
print("Total Expense Records:", len(expenses))
```

```python
# Total Spending
total_spent = expenses["Amount"].sum()
```

```python
category_spending = (
    df[df["Type"] == "Expense"]
    .groupby("Category")["Amount"]
    .sum()
    .sort_values(ascending=False)
)
print(category_spending)
```

```python
monthly_spending = (
    expenses.groupby(["Year", "Month"])["Amount"]
    .sum()
)

avg_monthly = monthly_spending.mean()

print("\n===== SPENDING ANALYSIS =====")
print("Total Spending:", round(total_spent, 2))
print("\nTop Categories:")
print(category_spending.head())

print("\nAverage Monthly Spending:", round(avg_monthly, 2))
```

```python
top_category = category_spending.idxmax()
print(top_category)
```

Anomaly detection

```python
mean_spend = monthly_spending.mean()
std_spend = monthly_spending.std()
print(f"Mean monthly spending: {mean_spend}")
print(f"Standard deviation of monthly spending: {std_spend}")
```

```python
upper_limit = mean_spend + 2.5 * std_spend
lower_limit = mean_spend - 2.5 * std_spend
```

```python
anomalies = monthly_spending[
    (monthly_spending > upper_limit) |
    (monthly_spending < lower_limit)
]
print("\n===== ANOMALY DETECTION =====")
print("Mean Monthly Spend:", round(mean_spend, 2))
print("Std Dev:", round(std_spend, 2))
print("Upper Limit:", round(upper_limit, 2))
print("Lower Limit:", round(lower_limit, 2))

if len(anomalies) > 0:
    print("\nAnomalies Found:")
    print(anomalies)
else:
    print("\nNo Anomalies Detected")
```
